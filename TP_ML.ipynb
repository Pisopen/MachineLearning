```python
ipAdress.dropna()
lower_bound_ip_address_int = []
for i in range(len(ipAdress['lower_bound_ip_address'])):
#     print(int(fraud['ip_address'][i]))
    lower_bound_ip_address_int.append(int(ipAdress['lower_bound_ip_address'][i]))
ipAdress['lower_bound_ip_address'] = lower_bound_ip_address_int
```

```python
ip_adress_int = []
for i in range(len(fraud['ip_address'])):
#     print(int(fraud['ip_address'][i]))
    ip_adress_int.append(int(fraud['ip_address'][i]))
fraud['ip_address'] = ip_adress_int
```
